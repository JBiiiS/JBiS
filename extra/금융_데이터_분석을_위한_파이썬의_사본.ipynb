###2. ndarray의 초기화

위에서는 리스트를 가지고 ndarray를 생성했지만, ndarray를 만드는 다양한 다른 방법이 존재합니다.

```python
# np.zeros((m,n)) : m by n null mtx
zero_mat = np.zeros((2,3))
print(zero_mat)
print('\n')

# np.ones((m,n)) : m by n mtx whose all entries are 1
one_mat = np.ones((2,3))
print(one_mat)
print('\n')

# np.full((m,n), c) : m by n whose all entries are c
constant_mat = np.full((2,3), 7)
print(constant_mat)
print('\n')

# np.eye(c) : Identity mtx whose dim = c
identity_mat = np.eye(3)
print(identity_mat)
print('\n')

# np.random.random((m,n)) : m by n whose enries are decided randomly in [0,1]
random_mat = np.random.random((2,2)) # 임의의 값으로 채워진 배열 생성
print(random_mat)

```

###6. Numpy 정수 인덱싱(integer indexing)

슬라이싱을 사용하면 배열로부터 부분 배열을 추출할 수 있지만, 연속적이지 않은 원소로 배열을 만들 경우에는 슬라이싱으로는 만들 수 없습니다.

예를 들어서 2행 2열의 원소와 5행 5열의 원소를 뽑아서 하나의 배열로 만들고자 하는 경우가 그렇습니다.

 이런 경우에는 인덱싱을 사용하여 배열을 구성할 수 있습니다. 인덱싱은 원하는 위치의 원소들을 뽑을 수 있습니다.

 결론: Numpy Indexing은 특정 원소들의 위치를 토대로, 그 원소들로 이루어진 vec 제작 가능

```python
mat = np.array([[1, 2], [4, 5], [7, 8]])
print(mat)
print('\n')

# 2행 1열의 원소
# => 0부터 카운트하므로 [1,0]의 원소.
print(mat[1, 0])
print('\n')

# mat[[2행, 1행],[0열, 1열]] 주의: 하나하나가 아니라, 앞 리스트에 순서대로 몇행인지, 뒤 리스트에 순서대로 몇열인지 쓰는거
# 즉, a31,a22를 가져와서 vec 제작해라
indexing_mat = mat[[2, 1],[0, 1]]
print(indexing_mat)
```

```python
#Without class, 계산기(메모리 기능 탑재) 함수를 제작하면,

result1 = 0
result2 = 0

def add1(num):  # 계산기1
    global result1 #global: 전역변수(함수의 계산 결과값을 유지 + 저장하는 변수)
    result1 += num
    return result1

def add2(num):  # 계산기2
    global result2
    result2 += num
    return result2

print(add1(3))
print(add1(4))
print(add2(3))
print(add2(7))

'''
와 같이 서로 다른 함수를 사용하고, 또 그 서로 다른 함수에 내가 적재적소에
잘 기억해서 숫자를 넣어줘야함. 근데 그게 한두개야 가능하지만 더 늘어나면?
'''
```

```python
#With class, making calculator program


class Calculator:
  def __init__(self):
    self.result = 0

  def add(self, num):
    self.result += num
    return self.result

cal1 = Calculator()
cal2 = Calculator()

print(cal1.add(3))
print(cal1.add(4))
print(cal2.add(3))
print(cal2.add(7))
```

바로 여기서 'Calculator'가 **class(클래스)**,

'cal1', 'cal2'가 **object(객체)**이다.

또한, 클래스 내 함수, 즉, 해당 예시 내 init, add는 **method(메서드)**라고 부른다.

객체들은 서로 독립, 즉, 영향을 미치지 않으므로, 다른 계산이 하고 싶으면 그냥 다른 객체 하나 만들면 된다.

```python
# 변환(conversion)
#strptime() : str 타입으로 된 날짜 데이터를 datetime 형태로 변환
#strftime() : datetime 형태의 데이터를 str 타입으로 변환
#주의해야 할 점은 우리가 형식을 정의할 때 지정한 형식이
#대소문자를 구분하므로 형식을 맞추거나 정확한 날짜를 표현해야 합니다.

'''
실제 분석에선 **문자열** 형태로 날짜/시간 주어지는 경우가 多
문자열로 주어진 시간을 자유자재로 datetime형태로 변환할 수 있어야함.

'''
#strptime function: str  ->  datetime

import datetime

datetime_str = '2018-05-13 12:34:56'

format = '%Y-%m-%d %H:%M:%S' #반드시 대소문자 지켜야함.
datetime_dt = datetime.datetime.strptime(datetime_str, format)
print(datetime_dt)
print(type(datetime_dt))
```

```python
#strftime function : datetime -> str

datetime_str = datetime_dt.strftime('%Y-%m-%d %H:%M:%S' )

print(type(datetime_str))
print(datetime_str)

datetime_str_part = datetime_dt.strftime('%Y-%m-%d %H')
print(type(datetime_str_part))
print(datetime_str_part)
```

### 날짜와 시간을 다루는 넘파이(Numpy)

**넘파이에서는 날짜와 시간을 datetime64 객체로 표현합니다.** 넘파이의 datetime64와 파이썬의 datetime.datetime과 차이점은 datetime에서는 시간을 마이크로초(10^-6) 단위까지 관리한다면, 넘파이의 datetime64는 아토초(10^-18) 단위까지 관리한다는 것입니다.

넘파이의 날짜와 시간을 이해해야 하는 가장 중용한 이유는 우리가 궁극적으로 사용할 **판다스의 날짜와 시간 함수가 넘파이를 기반으로 만들어졌기 때문**입니다. 그럼 넘파이의 날짜와 시간 사용 예제를 살펴봅시다.

#### datetime64

1. 문자열 형식으로 전달해서 생성
2. 유닉스 시간 이용

  첫번째 입력 매개변수는 수치 데이터, 그리고 두번째 입력 매개변수는 단위를 의미합니다.
  
  'ns'는 나노초 단위,
  
  'ms'는 밀리초 단위,
  
  's'는 초 단위,
  
  'D'는 일 단위
  
  를 의미합니다. 출력 결과는 단위에 따라서 각기 다른 종류의 날짜가 만들어집니다.

```python
import numpy as np
print(np.datetime64('2019-01-01'))
print('-'*20)

print(np.datetime64(1000,'ns'))
print('-'*20)

print(np.datetime64(10000000000,'ns'))
print('-'*20)

print(np.datetime64(10000000000,'ms'))
print('-'*20)

print(np.datetime64(10000000000,'s'))
print('-'*20)

print(np.datetime64(10000,'D'))

```

```python
# ret90_simple: 90개 데일리 단순 수익률
plt.figure(figsize=(12, 6))

# (1) 수익률 시계열 그래프
plt.subplot(2, 1, 1) #plt.subplot(행 개수, 열 개수, 몇번째 그림으로 할지)
plt.plot(ret90_simple.index, ret90_simple.values, color='blue', marker='o', markersize=3)
plt.axhline(0, color='gray', linestyle='--', linewidth=1)
plt.title(f'Daily Simple Returns (Last 90 Days from {ret90_simple.index[-1]})')
plt.ylabel('Return')
plt.grid(alpha=0.3)

# (2) 히스토그램 (분포)
plt.subplot(2, 1, 2)
plt.hist(ret90_simple, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Distribution of Daily Returns')
plt.xlabel('Daily Return')
plt.ylabel('Frequency')
plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()
ret90_simple.describe()


```

```python
# ret90_simple: 길이 90의 pandas Series (단순 수익률)
# NaN 제거 및 값 배열화
sample = ret90_simple.dropna().values
assert sample.size >= 1, "ret90_simple 비어있음"

# 파라미터
n = 90           # 표본 크기 (90일)
B = 3000         # 부트스트랩 반복 횟수
rng = np.random.default_rng(42)  # 재현성
'''
괄호안 숫자는 머가 들어와도 ㄱㅊ, 그냥 각기다른 난수발생기 상태를 만듦. 관습적으로 42
숫자를 넣기만 하면 앞seed 고정한 효과가 됨 즉,
default_rng()는 seed()의 발전된 형태임.
'''

# (방법 1) 벡터화: (B, n)로 한 번에 뽑고 평균
draws = rng.choice(sample, size=(B, n), replace=True) #sample내에서, 복원추출로 3000 x 90 mtx 제작
bootstrap_means = draws.mean(axis=1)
print(type(bootstrap_means))

# 확인
print(len(bootstrap_means))  # 3000
print(pd.Series(bootstrap_means).describe())

# bootstrap_means: 3000개 표본평균 (수익률 분포)
def calc_var(returns, conf=95):
    """
    returns : np.array 또는 pandas.Series (수익률 분포)
    conf    : 신뢰수준 (예: 95, 99 등)
    """
    alpha = 100 - conf  # 하위 (1-α)% 분위수
    var = np.percentile(returns, alpha)
    return var

# 예시
VaR_95 = calc_var(bootstrap_means, conf=95)
VaR_99 = calc_var(bootstrap_means, conf=99)

print(f"VaR(95%) = {VaR_95:.5f}")
print(f"VaR(99%) = {VaR_99:.5f}")
```

AND나 OR 또는 NOT과 같은 연산자를 사용하여 여러 개의 조건을 동시에 사용하거나, 조건문 자체를 반대로 해석하도록 할 수도 있습니다.

**데이터프레임에서는 AND, OR, NOT은 각각 &, |, ~로 표현합니다.**

**& : AND**

**| : OR**

**~ : NOT**

다시 말해 다음과 같이 사용할 수 있습니다.

A조건 & B조건 : A조건과 B조건 모두 만족하는 경우

A조건 | B조건 : A조건 또는 B조건 둘 중 하나를 만족하는 경우

~A조건 : A조건을 만족하는 경우의 반대. 즉, A조건을 만족하지 않는 경우.

물론 실제로는 조건이 2개 이상일 수도 있고, 이들을 섞어서 사용하는 것도 가능하므로 위 케이스보다 다양하게 사용할 수 있습니다. 여기서는 가장 기본적인 위의 세 가지 경우에 대해서만 다뤄보겠습니다.

```python
#NOT 조건
drink_df[~(drink_df.continent == 'EU')]
```

```python
#AND 조건
drink_df[(drink_df['continent'] == 'EU') & (drink_df.wine_servings > 300)]
```

```python
#OR 조건
drink_df[(drink_df['continent'] == 'EU') | (drink_df.wine_servings > 300)]
```

```python
print(len(drink_df[(drink_df['continent'] == 'EU') | (drink_df.wine_servings > 300)]))
print(len(drink_df[(drink_df.continent == 'EU') & (drink_df.wine_servings > 300)]))
```

##텍스트 마이닝을 위한 지식 : 토큰화


텍스트 분석에서 크롤링 등으로 얻어낸 코퍼스(말뭉치, 분석을 위해 수집한 원 언어 데이터) 데이터가 필요에 맞게 전처리되지 않은 상태라면, 해당 데이터를 사용하고자하는 용도에 맞게 토큰화(tokenization)하는 일을 하게 됩니다.

**주어진 코퍼스(corpus)에서 토큰(token)이라 불리는 단위로 나누는 작업을 토큰화(tokenization)** 라고 합니다. 토큰의 단위가 상황에 따라 다르지만, 보통 의미있는 단위로 토큰을 정의합니다. 예를 들어 단어 단위로 문장을 토큰화한다면 단어 토큰화라고 부릅니다. 한국어에서는 주로 형태소 분석기라는 것을 사용하여 단어 토큰화를 수행합니다.

한국어 형태소 분석기 패키지의 이름은 코엔엘파이로 다음과 같이 설치합니다.

###단어 토큰화

한국어 텍스트 분석에서 형태소 분석기를 사용하여 단어 토큰화를 수행해보겠습니다. 여기서는 Okt와 꼬꼬마 두 개의 형태소 분석기를 사용하여 토큰화를 수행하겠습니다.

```python
from konlpy.tag import Okt
from konlpy.tag import Kkma

okt = Okt()
kkma = Kkma()

print('OKT 형태소 분석 :', okt.morphs('열심히 코딩한 당신, 연휴에는 여행을 가봐요'))
print('OKT 품사 태깅 :', okt.pos('열심히 코딩한 당신, 연휴에는 여행을 가봐요'))
print('OKT 명사 추출 :'okt.nouns('열심히 코딩한 당신, 연휴에는 여행을 가봐요'))
```

위의 예제는 Okt 형태소 분석기로 토큰화를 시도해본 예제입니다. 각각의 메소드는 아래와 같은 기능을 갖고 있습니다.

1) morphs : 형태소 추출 (단어 토큰화)

2) pos : 품사 태깅(Part-of-speech tagging)

3) nouns : 명사 추출

Okt나 꼬꼬마 외의 코엔엘파이의 형태소 분석기들은 공통적으로 이 메소드들을 제공하고 있습니다. 위 예제에서 형태소 추출과 품사 태깅 메소드의 결과를 보면 조사를 기본적으로 분리하고 있음을 확인할 수 있습니다. 한국어 텍스트 분석에서 전처리에 형태소 분석기를 사용하는 것은 굉장히 유용합니다. 이번에는 꼬꼬마 형태소 분석기를 사용하여 같은 문장에 대해서 토큰화를 진행해봅시다

```python
print('꼬꼬마 형태소 분석 :',kkma.morphs("열심히 코딩한 당신, 연휴에는 여행을 가봐요"))
print('꼬꼬마 품사 태깅 :',kkma.pos("열심히 코딩한 당신, 연휴에는 여행을 가봐요"))
print('꼬꼬마 명사 추출 :',kkma.nouns("열심히 코딩한 당신, 연휴에는 여행을 가봐요"))
```

앞서 사용한 Okt 형태소 분석기와 결과가 다른 것을 볼 수 있습니다. 각 형태소 분석기는 성능과 결과가 다르게 나오기 때문에, 형태소 분석기의 선택은 사용하고자 하는 필요 용도에 어떤 형태소 분석기가 가장 적절한지를 판단하고 사용하면 됩니다.

## 텍스트 마이닝을 위한 지식 : 뉴스 본문 추출

```python
def scrape_article_content(url):
    """
    주어진 URL에서 뉴스 기사의 제목과 본문을 추출합니다.

    Args:
        url (str): 추출할 뉴스 기사의 URL.

    Returns:
        dict: 'title'과 'content' 키를 가진 딕셔너리. 추출 실패 시 None 반환.
    """

    # 1. URL에 접속 및 HTML 가져오기
    try:
        # User-Agent를 설정하여 봇 접근을 숨기고 차단을 피합니다.
        headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}
        response = requests.get(url, headers=headers, timeout=10)
        response.raise_for_status() # HTTP 오류 발생 시 예외 발생

        # 인코딩 문제 방지
        response.encoding = response.apparent_encoding #response.apparent_encoding은 자동적으로 encode할 language 결정하는 명령어
        html = response.text

    except requests.exceptions.RequestException as e:
        print(f"Error accessing URL {url}: {e}")
        return None

    # 2. BeautifulSoup을 사용하여 HTML 파싱
    soup = BeautifulSoup(html, 'html.parser')

# 3. 제목 추출
    # 일반적인 뉴스 기사의 제목 태그를 우선순위로 시도합니다.
    # 메타 태그의 Open Graph 제목을 먼저 시도하는 것이 안정적일 수 있습니다.
    title = None
    try:
        # 1순위: Open Graph Title
        title_tag = soup.find("meta", property="og:title")
        if title_tag and title_tag.get("content"):
            title = title_tag["content"].strip()

        # 2순위: <title> 태그 (og:title이 없을 경우)
        if not title:
            title_tag = soup.find("title")
            if title_tag:
                title = title_tag.get_text().strip()
                # 불필요한 부분(예: ' - 뉴스 채널명') 제거 (선택 사항)
                if ' - ' in title:
                    title = title.split(' - ')[0]

    except Exception as e:
        print(f"Error extracting title: {e}")
        title = "제목 추출 실패"

    # 4. 본문 추출 및 정리
    content = []

    # **핵심:** 뉴스 기사 본문이 포함될 가능성이 높은 태그/클래스를 지정합니다.
    # 이 부분은 **웹사이트마다 커스터마이징이 필요**한 가장 중요한 부분입니다.
    # 아래는 일반적인 기사 본문 영역의 클래스/ID 예시입니다.

    # 예시: 기사 본문이 <div id="articleBody"> 또는 <div class="news_content"> 안에 있는 경우
    possible_body_selectors = [
        'div[id*="articleBody"]',
        'div[class*="content-body"]',
        'div[class*="article-view-content"]',
        'div[class*="news_content"]',
        'div[class*="entry-content"]',
    ]

    article_body_container = None
    for selector in possible_body_selectors:
        if selector.startswith('div[id*='):
            # ID가 포함된 div 찾기
            id_part = selector.split('"')[1]
            article_body_container = soup.find('div', id=lambda x: x and id_part in x)
        elif selector.startswith('div[class*='):
            # Class가 포함된 div 찾기
            class_part = selector.split('"')[1]
            article_body_container = soup.find('div', class_=lambda x: x and class_part in x)

        if article_body_container:

            break
    print(article_body_container)
    # 컨테이너에서 <p> 태그의 텍스트를 모두 추출
    if article_body_container:
        paragraphs = article_body_container.find_all('p')
        for p in paragraphs:
            text = p.get_text().strip()
            if text:
                content.append(text)
    print(content)
    # 컨테이너를 찾지 못했거나 <p> 태그가 없는 경우, 모든 텍스트를 시도
    if not content:
        # 본문 영역을 찾지 못했다면, 일반적인 텍스트 영역을 시도
        # 모든 텍스트를 가져온 후 불필요한 공백을 제거
        main_content_tags = soup.find_all(['p', 'div', 'span'], class_=lambda x: x and ('content' in x or 'body' in x))

        if main_content_tags:
             # 가장 긴 텍스트를 본문으로 가정합니다.
            full_text = ' '.join([tag.get_text().strip() for tag in main_content_tags if len(tag.get_text().strip()) > 300])
            content.append(full_text)
        elif article_body_container:
            # 컨테이너는 찾았는데 P 태그가 없는 경우, 컨테이너 내의 모든 텍스트를 시도
            content.append(article_body_container.get_text().strip())
        else:
            # 최종 실패 시 메시지
            content.append("본문 추출 실패: 적절한 본문 컨테이너를 찾을 수 없습니다. 셀렉터 조정이 필요합니다.")


    # 추출된 본문을 하나의 문자열로 결합하고 정리
    # 여러 줄 공백을 한 줄 공백으로 바꾸고, 줄바꿈으로 문단 구분
    final_content = '\n\n'.join(content).strip()

    return {
        'title': title,
        'content': final_content
    }

# --- 사용 예시 ---
# **주의: 실제로 작동하는 뉴스 기사 URL로 대체해야 합니다.**
# (예시 URL은 실제 접속 시 에러가 발생할 수 있습니다.)
# 실제 테스트를 위해 네이버 뉴스나 다음 뉴스 등의 URL을 사용해 보세요.
article_url = "https://n.news.naver.com/mnews/article/018/0004960169"

print(f"--- 기사 URL: {article_url} ---")
scraped_data = scrape_article_content(article_url)

if scraped_data:
    print("\n✅ 추출 완료")
    print("-----------------------------------")
    print(f"**제목:** {scraped_data['title']}")
    print("-----------------------------------")
    print("**본문:**")
    print(scraped_data['content'])
    #+[:500] '...' if len(scraped_data['content']) > 500 else scraped_data['content']
    print("-----------------------------------")
```

###자동 미분

tape.gradient()는 자동 미분 기능을 수행합니다.

```python
import tensorflow as tf

# 미분 대상 변수 설정 (초기값 2)
w = tf.Variable(2.)

# 함수 정의: z = 2w^2 + 5
def f(w):
    y = w**2          # y = w^2
    z = 2*y + 5       # z = 2w^2 + 5
    return z

# 자동 미분을 위한 GradientTape
with tf.GradientTape() as tape:
    z = f(w)          # 연산 기록 (z가 w에 어떻게 의존하는지)
                      # chain rule로 (dz/dy)(dy/dw)과정이구나! 하고 있음

# dz/dw 계산
gradients = tape.gradient(z, [w]) #[w]로 쓴건 2개 이상 변수에 대해 단변수 defernetiate 가능해서임.

# 미분 결과 출력 (w=2 → 4w = 8)
print(gradients)
```

###자동 미분을 이용한 선형 회귀 구현

```python
w = tf.Variable(4.0)
b = tf.Variable(1.0)

@tf.function #파이썬 함수를 TensorFlow 연산 그래프로 변환 (graph execution), 이렇게 하면 tf가 하나로 묶어 연산해서 속도가 빨라짐
def hypothesis(x):
    return w * x + b

x_test = [3.5,5,5.5,6]
print(hypothesis(x_test).numpy())
```

```python
@tf.function
def mse_loss(y_pred, y):
  return tf.reduce_mean(tf.square(y_pred - y))
```

```python
x = [1, 2, 3, 4, 5, 6, 7, 8, 9] # 공부하는 시간
y = [11, 22, 33, 44, 53, 66, 77, 87, 95] # 각 공부하는 시간에 맵핑되는 성적

optimizer = tf.optimizers.SGD(0.001) #옵티마이저로 SGD(경사하강법) 사용, learning rate = 0.01

for i in range(301):
  with tf.GradientTape() as tape:
    y_pred = hypothesis(x)
    cost = mse_loss(y_pred, y)

  gradients = tape.gradient(cost,[w,b])

  optimizer.apply_gradients(zip(gradients,[w,b])) # zip(gradients, [w,b]) = [[gradeints_w, w], [gradients_b, b]]

  if i % 10 == 0:
    print("epoch : {:3} | w의 값 : {:5.4f} | b의 값 : {:5.4} | cost : {:5.6f}".format(i, w.numpy(), b.numpy(), cost)) #{5.4f}에서 전자의 number는 minimum width,
                                                                                                                      # ex) -12.12345 -> -12.12345, -1.23 -> -1.2300, print된 값이 잘 arrange되게 끔.


```

##단층 퍼셉트론

단층 퍼셉트론은 값을 보내는 단계과 값을 받아서 출력하는 두 단계로만 이루어집니다. 이때 **이 각 단계를 보통 층(layer)이라고 부르며, 이 두 개의 층을 입력층(input layer)과 출력층(output layer)이라고 합니다**

 **컴퓨터는 두 개의 값 0과 1을 입력해 하나의 값을 출력하는 회로가 모여 만들어지는데, 이 회로를 게이트(gate)라고 부릅니다.**

 ex)

 AND gate: 두개의 Input이 1일 때만 1 산출

NAND gate: AND gate 정반대

 OR gate: 하나 이상의 Input이 1이면 1 산출

##다층 퍼셉트론 (MultiLayer Perceptron, MLP)

다층 퍼셉트론과 단층 퍼셉트론의 차이는 단층 퍼셉트론은 입력층과 출력층만 존재하지만, 다층 퍼셉트론은 중간에 층을 더 추가하였다는 점입니다. 이렇게 **입력층과 출력층 사이에 존재하는 층을 은닉층(hidden layer)이라고 합니다.**



```python
def NAND_gate(x1, x2):
    w1 = -0.5
    w2 = -0.5
    b = 0.7
    result = x1*w1 + x2*w2 + b
    if result <= 0:
        return 0
    else:
        return 1


def OR_gate(x1, x2):
    w1 = 0.6
    w2 = 0.6
    b = -0.5
    result = x1*w1 + x2*w2 + b
    if result <= 0:
        return 0
    else:
        return 1

def XOR_MLP(x1, x2):
    h1 = NAND_gate(x1, x2)  # (1,1) 제거
    h2 = OR_gate(x1, x2)    # (0,0) 제거
    if h1 == 0 or h2 == 0:
        return 0
    else:
        return 1


for x1, x2 in [(0,0),(0,1),(1,0),(1,1)]:
    print(x1, x2, XOR_MLP(x1, x2))

```

Sigmoid Function(시그모이드 함수)

image.png

인공 신경망의 학습 과정은 다음과 같습니다. 우선 인공 신경망은 입력에 대해서 순전파(forward propagation) 연산을 하고, 그리고 순전파 연산을 통해 나온 예측값과 실제값의 오차를 손실 함수(loss function)을 통해 계산하고, 그리고 이 손실(오차라고도 부릅니다. loss)을 미분을 통해서 기울기(gradient)를 구하고, 이를 통해 출력층에서 입력층 방향으로 가중치와 편향을 업데이트 하는 과정인 역전파(back propagation)를 수행합니다. 역전파에 대해서는 뒤에서 더 자세히 설명하겠지만 일단 여기에서는 인공 신경망에서 출력층에서 입력층 방향으로 가중치와 편향을 업데이트 하는 과정이라고만 언급해두겠습니다. 역전파 과정에서 인공 신경망은 경사 하강법을 사용합니다.

이 시그모이드 함수의 문제점은 미분을 해서 기울기(gradient)를 구할 때 발생합니다.

image.png

주황색 구간에서는 미분값이 0에 가까운 아주 작은 값입니다. 초록색 구간에서의 미분값은 최대값이 0.25입니다. 다시 말해 시그모이드 함수를 미분한 값은 적어도 0.25 이하의 값입니다. 시그모이드 함수를 활성화 함수로하는 인공 신경망의 층을 쌓는다면, 가중치와 편향을 업데이트 하는 과정인 역전파 과정에서 0에 가까운 값이 누적해서 곱해지게 되면서, 앞단에는 기울기(미분값)가 잘 전달되지 않게 됩니다. 이러한 현상을 기울기 소실(Vanishing Gradient) 문제라고 합니다.

시그모이드 함수를 사용하는 은닉층의 개수가 다수가 될 경우에는 0에 가까운 기울기가 계속 곱해지면 앞단에서는 거의 기울기를 전파받을 수 없게 됩니다. 다시 말해 매개변수 w가 업데이트 되지 않아 학습이 되지를 않습니다.

image.png

**결론적으로 시그모이드 함수의 은닉층에서의 사용은 지양됩니다. 시그모이드 함수는 주로 이진 분류를 위해 출력층에서 사용합니다.**

```python
# 시그모이드 함수 그래프를 그리는 코드
def sigmoid(x):
    return 1/(1+np.exp(-x))
'''
위 함수가 sigmoid function, 쓰임은 '변수의 변형'임.
예컨데, 앞선 예시의 wx+b와 같은 regression equation이 있다 치면,
σ(wx+b) = 1/(1+exp(-(wx+b)))
이 꼴로 만드는 역할.
애초의 original regression 자체를 reform하니깐, 당연히 도함수들도
영향을 받는거고, vanishing이 generated됨.
'''

x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)

plt.plot(x, y)
plt.plot([0,0],[1.0,0.0], ':') # 가운데 점선 추가
plt.title('Sigmoid Function')
plt.show()
```

####Hyperbolic Tangent Function(하이퍼볼릭탄젠트 함수)

**하이퍼볼릭탄젠트 함수(tanh)는 입력값을 -1과 1사이의 값으로 변환합니다. 그래프를 그려보겠습니다.**

하이퍼볼릭탄젠트 함수도 -1과 1에 가까운 출력값을 출력할 때, 시그모이드 함수와 같은 문제가 발생합니다. 그러나 하이퍼볼릭탄젠트 함수의 경우에는 시그모이드 함수와는 달리 0을 중심으로 하고있으며 **하이퍼볼릭탄젠트 함수를 미분했을 때의 최대값은 1로 시그모이드 함수의 최대값인 0.25보다는 큽니다.** 다시 말해 미분했을 때 시그모이드 함수보다는 전반적으로 큰 값이 나오게 됩니다. 그래서 **시그모이드 함수보다는 기울기 소실 증상이 적은 편이며 은닉층에서 시그모이드 함수보다는 선호됩니다.**

```python
x = np.arange(-5.0, 5.0, 0.1) # -5.0부터 5.0까지 0.1 간격 생성
y = np.tanh(x)

plt.plot(x, y)
plt.plot([0,0],[1.0,-1.0], ':')
plt.axhline(y=0, color='orange', linestyle='--')
plt.title('Tanh Function')
plt.show()
```

####ReLU Function(렐루 함수)

인공 신경망의 은닉층에서 가장 인기있는 함수입니다. 수식은
image.png
로 아주 간단합니다.

렐루 함수는 음수를 입력하면 0을 출력하고, 양수를 입력하면 입력값을 그대로 반환하는 것이 특징인 함수로 출력값이 특정 양수값에 수렴하지 않습니다. 0이상의 입력값의 경우에는 미분값이 항상 1입니다. 깊은 신경망의 은닉층에서 시그모이드 함수보다 훨씬 더 잘 작동합니다. 뿐만 아니라, 렐루 함수는 시그모이드 함수와 하이퍼볼릭탄젠트 함수와 같이 어떤 연산이 필요한 것이 아니라 단순 임계값이므로 연산 속도도 빠릅니다.

하지만 여전히 문제점이 존재하는데, **입력값이 음수면 기울기. 즉, 미분값도 0이 됩니다. 그리고 이 뉴런은 다시 회생하는 것이 매우 어렵습니다. 이 문제를 죽은 렐루(dying ReLU)라고 합니다.**

```python
def relu(x):
    return np.maximum(0, x)

x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)

plt.plot(x, y)
plt.plot([0,0],[5.0,0.0], ':')
plt.title('Relu Function')
plt.show()
```

####Leaky ReLU(리키 렐루 함수)
죽은 렐루를 보완하기 위해 ReLU의 변형 함수들이 등장하기 시작했습니다. 변형 함수는 여러 개가 있지만 여기서는 Leaky ReLU에 대해서만 소개합니다. Leaky ReLU는 입력값이 음수일 경우에 0이 아니라 0.001과 같은 매우 작은 수를 반환하도록 되어있습니다. 수식은
로 아주 간단합니다. a는

image.png

하이퍼파라미터로 Leaky('새는') 정도를 결정하며 일반적으로는 0.01의 값을 가집니다. 여기서 말하는 '새는 정도'라는 것은 입력값의 음수일 때의 기울기를 비유하고 있습니다.



```python
a = 0.1

def leaky_relu(x):
    return np.maximum(a*x, x)

x = np.arange(-5.0, 5.0, 0.1)
y = leaky_relu(x)

plt.plot(x, y)
plt.plot([0,0],[5.0,0.0], ':')
plt.title('Leaky ReLU Function')
plt.show()
```

####Softmax Function(소프트맥스 함수)

image.png


**은닉층에서는 ReLU(또는 ReLU 변형) 함수들을 사용하는 것이 일반적입니다. 반면, 소프트맥스 함수는 시그모이드 함수처럼 출력층에서 주로 사용됩니다. 시그모이드 함수가 두 가지 선택지 중 하나를 고르는 이진 분류 (Binary Classification) 문제에 사용된다면 소프트맥스 함수는 세 가지 이상의 (상호 배타적인) 선택지 중 하나를 고르는 다중 클래스 분류(MultiClass Classification) 문제에 주로 사용됩니다.** 다시 말해서 딥 러닝으로 이진 분류를 할 때는 출력층에 앞서 배운 로지스틱 회귀를 사용하고, 딥 러닝으로 다중 클래스 분류 문제를 풀 때는 출력층에 소프트맥스 회귀를 사용한다고 생각할 수 있습니다.

```python
# @title
x = np.arange(-5.0, 5.0, 0.1) # -5.0부터 5.0까지 0.1 간격 생성
y = np.exp(x) / np.sum(np.exp(x))

plt.plot(x, y)
plt.title('Softmax Function')
plt.show()
```

총 2,295개의 행과 4개의 열로 구성되어져 있습니다. 첫번째 열은 종목 번호에 해당하는 code 열, 두번째 열은 해당 종목이 KOSPI인지 KOSDAQ인지를 알려주는 market 열, 세번째 열은 회사명에 해당하는 name 열, 그리고 네번째 business 열은 우리가 학습할 사업 보고서에 해당됩니다.

이제 business 열에 대해서 형태소 분석을 수행해봅시다. 이와 동시에 Doc2Vec 학습을 위해서 필요한 형식으로 데이터의 형식을 변환합니다. Doc2Vec 학습을 위해서는 해당 문서의 '제목'과 단어 토큰화가 된 상태의 해당 문서의 '본문' 두 가지가 필요합니다.

TaggedDocument의 tags에 해당 문서의 '제목'을, 그리고 words에 해당 문서의 '본문'에 해당하는 단어 토큰화 결과를 저장하고, 이 결과를 원소로 하는 파이썬 리스트인 tagged_corpus_list를 만듭니다.

```python
mecab = Mecab()

tagged_corpus_list = []

for index, row in tqdm(df.iterrows(), total = len(df)):
  text = row['business']
  tag = row['name']
  tagged_corpus_list.append(TaggedDocument(words = mecab.morphs(text),
                                           tags = [tag]))

print('문서의 수', len(tagged_corpus_list))

'''
explanation on codes

iterrows()는 DataFrame의 method -> 행 단위로 쪼개줌. index는 index, row는 pd.Seris로

mecab : '형태소' 분석기
mecab.morphs(text): text내에 있는 corpus를, 형태소 단위로 쪼개라.
-> raw tokenization step. 즉, 불용어 제거나 noun extraction 과 같은 process는 omitted

tqdm: for문에서 진행률(progress bar) 표시용에 불과. 출력된 거 보면 이새끼 역할 알임

**gensim은 “대용량 텍스트에서 의미(semantic structure)를 학습하는
비지도 NLP 모델들을 구현해 둔 파이썬 라이브러리”**

word2vec, doc2vec과 같은 Unsupervised Learning / Self-Supervised Learning
모델들이 들어가 있음.

그중 "TaggedDocument"는, 문서 벡터를 학습하기 위한 학습용 데이터를 만드는 명령어
words : 전체 'document'를 morphs/words단위로 쪼개준거 / tags: the name(tag) of each document
로 해서 각각의 문서를 list에 넣어주고 있는 거임.

visualizingL tagged_corpus_list = [D_1,...,D_n]
where D_i = TaggedDocument(morphs_i(business_i), tag_i = name_i)
'''

```

##Doc2Vec 학습 및 테스트



```python
from gensim.models import doc2vec
```

```python
model = doc2vec.Doc2Vec(vector_size=300, alpha=0.025, min_alpha = 0.025,
                        workers= 8, window= 8)
'''
Doc2Vec

vector_size: 문서 벡터의 차원
즉 각 기업이 300차원의 벡터화됨.
(보통, 100 is small, 300 is standard, and more than 500 are many(so are used when the data are big))

alpha: the initial learning rate
min_alpha: the final learning rate
(In this example, the alpha is equal to min_alpha, which means there isn't any lr decay, the fixed lr,
this kind of settings are usually exploited when the data size is relatively small and the converging speed is slow)
(the basically set values are alpha = 0.025, min_alpha = 0.00001)

workers: the number of cores(cpu, not gpu)

window: the maximum distance between the target word and its neighboring word
(In this example, window = 8 means the before and after 8 words are viewed as a context)
(If the window's value is small, the sequence is exploited to verify the gramatical resemblance, if big, then semantic one.)

build_vocab():the kind of step that constructs the prior settings where
takes a scan of all words, calculates the frequency of words, and so on.

total_examples in train : used to make the process of diminishing lr and calculate the progress internally in gensim

epoch is usually assigned into the interval [10,30], generally 20
'''

# Vocabulary 빌드
model.build_vocab(tagged_corpus_list)

# Doc2Vec 학습
model.train(tagged_corpus_list, total_examples=model.corpus_count, epochs=20)

#save model
model.save('dart.doc2vec')
```

PR Curve: Precision(y-ax) - Recall(x-ax)

즉 recall을 너무 올리다 보면 precision과의 tradeoff가 커져 어느순간 너무 precision 급락 시작

```python
from sklearn.metrics import precision_recall_curve, confusion_matrix

# 1) val에서 예측확률 뽑기
p_val = model.predict(X_val_np).ravel()

# 2) PR curve로 threshold 후보별 precision/recall 계산
precision, recall, thresholds = precision_recall_curve(y_val_np, p_val)
# precision/recall은 thresholds보다 길이가 1 더 김 (맨끝은 threshold=inf 취급)
precision, recall = precision[:-1], recall[:-1]

'''
ravel(): output dimension이 1이 아닐 수 있기에, 기본적으로 model.predict는
np.array([[1,2],[1,2]])처럼 .ndim = 2, .shape(n,m>=1)로 나옴. 즉,
여기서는 w/o ravel()이었으면, p_val.ndim = 2, p_val.shape = (n, 1)로 나오는데,
ravel()이 붙으면 1차원 mtx를 vec화 해줌. p_val.ndim = 1, p_val.shape = (n,)

precision_recall_curve: 말그대로 PR Curve 만들어줌
만들 때, 맨마지막에 각각의 threshold값이 np.array들어오고, 그 앞에
해당 threshold에서의 precision & recall이 마찬가지로 np.array로 들어옴
'''


# 3) 목표 recall 설정
target_recall = 0.90

mask = recall >= target_recall
if not np.any(mask):
    print("val에서 target_recall을 만족하는 threshold가 없음. target을 낮추거나 모델/가중치 조정 필요.")
else:
    best_idx = np.argmax(precision[mask])           # precision 최대
    best_threshold = thresholds[mask][best_idx]
    best_precision = precision[mask][best_idx]
    best_recall = recall[mask][best_idx]

    print(f"Best threshold (recall≥{target_recall}): {best_threshold:.4f}")
    print(f"Precision: {best_precision:.4f} | Recall: {best_recall:.4f}")


'''
confusion mtx는 TP/FP/TN/FN 자동 구해주는 명령어. 순서는 하단 참조
'''

# 4) confusion matrix도 같이 확인
yhat_val = (p_val >= best_threshold).astype(int)
tn, fp, fn, tp = confusion_matrix(y_val_np, yhat_val).ravel()
print(f"TP={tp}, FP={fp}, FN={fn}, TN={tn}")
```

```python
import numpy as np
from sklearn.metrics import (
    confusion_matrix,
    precision_score,
    recall_score,
    f1_score,
    average_precision_score,
    roc_auc_score,
    classification_report,
)

# 1) test 예측확률
p_test = model.predict(X_test_np).ravel()

# 2) threshold 적용해서 0/1 예측
yhat_test = (p_test >= best_threshold).astype(int)

# 3) confusion matrix
tn, fp, fn, tp = confusion_matrix(y_test_np, yhat_test).ravel()

# 4) 핵심 지표 (threshold 기반)
prec = precision_score(y_test_np, yhat_test, zero_division=0)
rec  = recall_score(y_test_np, yhat_test, zero_division=0)
f1   = f1_score(y_test_np, yhat_test, zero_division=0)

# 5) threshold-독립 지표 (score 기반)
pr_auc  = average_precision_score(y_test_np, p_test)  # PR-AUC (Average Precision)
roc_auc = roc_auc_score(y_test_np, p_test)

print(f"[TEST] threshold = {best_threshold:.6f}")
print(f"TP={tp}, FP={fp}, FN={fn}, TN={tn}")
print(f"Precision={prec:.4f} | Recall={rec:.4f} | F1={f1:.4f}")
print(f"PR-AUC(AP)={pr_auc:.4f} | ROC-AUC={roc_auc:.4f}")

# 6) 상세 리포트(선택)
print("\nClassification report:")
print(classification_report(y_test_np, yhat_test, digits=4, zero_division=0))

# ---- (선택) 기본 threshold=0.5와 비교 ----
yhat_test_05 = (p_test >= 0.5).astype(int)
prec05 = precision_score(y_test_np, yhat_test_05, zero_division=0)
rec05  = recall_score(y_test_np, yhat_test_05, zero_division=0)
f105   = f1_score(y_test_np, yhat_test_05, zero_division=0)
tn05, fp05, fn05, tp05 = confusion_matrix(y_test_np, yhat_test_05).ravel()

print("\n[TEST] threshold = 0.5 (baseline)")
print(f"TP={tp05}, FP={fp05}, FN={fn05}, TN={tn05}")
print(f"Precision={prec05:.4f} | Recall={rec05:.4f} | F1={f105:.4f}")
```
